```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data: square feet (X) vs price in $1000s (y)
X = np.array([[650], [800], [900], [1000], [1200]])  # square footage
y = np.array([150, 200, 230, 250, 300])              # price in $1000s

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Predict
X_new = np.array([[1100]])
y_pred = model.predict(X_new)

print(f"Predicted price for 1100 sq ft: ${y_pred[0]*1000:.2f}")

# Plotting
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, model.predict(X), color='red', label='Linear Fit')
plt.xlabel('Square Feet')
plt.ylabel('Price ($1000s)')
plt.title('Linear Regression Example')
plt.legend()
plt.grid(True)
plt.show()
```
